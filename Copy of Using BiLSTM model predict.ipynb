```python
model = tf.keras.models.load_model('drive/My Drive/Colab Notebooks/Analytics Bidhya each step/analytics.h5')
```

```python
import pandas as pd
```

```python
df=pd.read_csv('drive/My Drive/Colab Notebooks/Analytics Bidhya each step/Datasets/combined reviews.csv')
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
num_words = 10000
```

```python
tokenizer = Tokenizer(num_words=num_words)
tokenizer.fit_on_texts(df['review'])
sequences = tokenizer.texts_to_sequences(df['review'])
X = pad_sequences(sequences, maxlen=300)
```

```python
predicted_emotions = model.predict(X)
```

```python
print(df.shape)
```

```python
import numpy as np
```

```python
df['predicted_emotions'] = predicted_emotions.argmax(axis=-1)
df['predicted_emotions'] = df.predicted_emotions.replace({0:'joy',1:'anger',2:'love',3:'sadness',4:'fear',5:'surprise'})
```

```python
df.to_csv('drive/My Drive/Colab Notebooks/Analytics Bidhya each step/Datasets/preprocessed_data_with_predictions.csv', index=False)
```
